# Pie chart

```python
#from England, Germany and Spain.

#Creating new columns that contain England, Germany, spain and Argentina and 
# one column labeled "other for all other nationalities:"

df.loc[df.Nationality =='England', 'Nationality_2'] = 'England'
df.loc[df.Nationality =='Germany', 'Nationality_2'] = 'Germany'
df.loc[df.Nationality =='Spain', 'Nationality_2'] = 'Spain'
df.loc[df.Nationality =='Argentina', 'Nationality_2'] = 'Argentina'
df.loc[~df.Nationality.isin(['England','Germany','Spain','Argentina']), 'Nationality_2'] = 'Others'

#Next, let's create a dictionary that contains the proportion values for each of these:

proportion = dict(Counter(df['Nationality_2']))
for key, values in proportion.items(): 
    proportion[key] = values/len(df)*100
print(proportion)
```

```python
#how to create percentages
plt.pie(proportion.values(),labels = proportion.keys(),autopct='%1.2f%%') # %1.2f%% 1 means atleast 1 number before decimal
#2 means 2 decimal values # last % is symbol after value

plt.figure(figsize =(10,7))
plt.show()

#Note: autopct enables you to display the percent value using python string formatting.
#for example, if autopct='%.2f', then for each pie wedge, the format string is "%.2f" and
#the numerical percent value for that wedge is pct,
# so the wedge label is set to the string '%.2f'%pct.
#
```

```python
explode = (0,0,0,0,0.3)
plt.pie(proportion.values(), labels=proportion.keys(), autopct='%1.2f%%', explode=explode)
plt.figure(figsize=(10,7))
plt.show()
```

# Seaborn Library

```python
# To regenerate our histogram of the overall column,
#we use the distplot method on the Seaborn object

import seaborn as sns
```

```python
# distribution plot show us the nature of the plot whether the distribution is normal distribution, right skewed or left skewed
sns.distplot(df['Overall'], color='green') #distribution plot
plt.xlabel('Overall')
plt.ylabel('Frequency')
plt.title('Histogram of Overall Rating')
plt.grid()
plt.show()
```

```python
#seaborn also makes generating scatter plots straightforward.
#Let's recreate the scatter plot from earlier:
sns.scatterplot(df['Overall'], df['Wage'])
plt.title('Overall vs. Wage')
plt.ylabel('Wage')
plt.xlabel('Overall')
plt.grid()
plt.show()
```

```python
# 6. Heat Maps

# Seaborn is also known for making correlation heatmaps, which can be used to identify variable dependence.
#To generate one, first we need to calculate the correlation between a set of numerical columns.
#Let's do this for age, overall, Wage and skill moves;

corr = df[['Overall','Age','Wage','Skill Moves']].corr() #calculating correlation with each other
sns.heatmap(corr, annot=True) # annot for annotate to true to see the correlation values
plt.title('Heatmap')
plt.show()
```

```python
# 7. Pair Plots
# This allows you to generate a matrix of distributions and scatter plots for a set of numerical features.
# It is a quick and easy way to visualize both the distribution  in numerical values and relationship
#between variables through scatterplots.
#let's do this for age, overall and potential:

data = df[['Overall','Age','Potential',]]
sns.pairplot(data)
plt.show()
```
